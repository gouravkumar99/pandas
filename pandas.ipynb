```python
#MERGING BY A COMMON KEY
```

```python
left = pd.DataFrame({'Key':['k0','k1','k2','k3'],
                    'A':['a0','a1','a2','a3'],
                     'B':['b0','b1','b2','b3']
                    })
left
```

```python
right = pd.DataFrame({'Key':['k0','k1','k2','k3'],
                    'A':['c0','c1','c2','c3'],
                     'B':['d0','d1','d2','d3']
                    })
right
```

```python
mrg = pd.merge(left,right,how='inner',on='Key')
print(mrg)
```

```python
data = {"firstname":['gourav',None,'Tina','Shivani','Janifer'],
       'lastname':['kumar',None,'sharma','gupta','martin'],
        'age':[42,None,36,24,73],
       'sex':['M',None,"F",'F','M'],
        'prescore':[4,None,None,2,3],
        'postscore':[25,None,None,62,7]
        
       }
```

```python
df = pd.DataFrame(data, columns = ['firstname','lastname','sex'])
df
```

```python
df= pd.DataFrame(data)
df
```

```python
#dropna use for drope nan rows it's like a filter

df.dropna()
```

```python
# if we use dropna(how='all') so only drop full rowvalues not columns
df.dropna(how='all')
```

```python
df['location'] = None
df
```

```python
df.dropna(axis=1,how='all')
df
```

```python
df.fillna(0)
```

```python
df
```

```python
df['prescore'].fillna(df['prescore'].mean(),inplace=True)
df
```

```python
df['postscore'].fillna(df.groupby('sex')['postscore'].transform('mean'),inplace=True)
df
```

```python
df[df['age'].notnull() & df['sex'].notnull()]
```

```python
df
```

```python
# back fill
```

```python
df.fillna(method='bfill')
```

```python
# forward fill
```

```python
df = df.fillna(method='ffill')
df
```

```python
# if we are use axis = 1 only flot and int.. no fill
df.fillna(method='ffill',axis=1)
```
